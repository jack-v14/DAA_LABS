```python
import time
import matplotlib.pyplot as plt
import random

# Recursive Binary Search
def binary_search_recursive(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search_recursive(arr, left, mid - 1, target)
    else:
        return binary_search_recursive(arr, mid + 1, right, target)

# Run the experiment
def run_experiment():
    sizes = [1000, 5000, 10000, 20000, 50000, 100000, 200000]
    times = []

    for n in sizes:
        arr = list(range(n))
        target = random.choice(arr)

        start = time.time()
        binary_search_recursive(arr, 0, n - 1, target)
        end = time.time()

        elapsed_time = (end - start) * 1000  # convert to milliseconds
        times.append(elapsed_time)
        print(f"n = {n}, time = {elapsed_time:.6f} ms")

    return sizes, times

# Plot the graph
def plot_graph(sizes, times):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, marker='o', color='blue')
    plt.title("Time Taken vs Array Size (Recursive Binary Search)")
    plt.xlabel("Array Size (n)")
    plt.ylabel("Time Taken (ms)")
    plt.grid(True)
    plt.show()

# Execute
sizes, times = run_experiment()
plot_graph(sizes, times)
```
